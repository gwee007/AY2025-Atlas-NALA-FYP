```python
from langchain_huggingface.embeddings import HuggingFaceEmbeddings
qwen_embeddings = HuggingFaceEmbeddings(model_name="Qwen/Qwen3-Embedding-0.6B")
```

```python
from langchain_experimental.text_splitter import SemanticChunker
text_splitter = SemanticChunker(qwen_embeddings) # use percentile breakpoint by default
```

```python
file_path = "../../data/processed_lectures/lecture3_laplace_transforms/cleaned_text_laplace_transform.txt"
string_to_remove = "CH3101 Chapter 3: Laplace Transforms"

# 1. Read the file content
try:
    with open(file_path, 'r', encoding="utf-8") as file:
        file_content = file.read()
    
    print(f"Length of file content before removal: {len(file_content)} characters")

    # 2. Perform the replacement
    modified_content = file_content.replace(string_to_remove, "")

    # remove all leading whitespace
    modified_content = modified_content.lstrip()
    
    # 3. Write the modified content back to the file
    with open(file_path, 'w', encoding="utf-8") as file:
        file.write(modified_content)

    print(f"Length of file content after removal: {len(modified_content)} characters")
    print(f"Successfully removed '{string_to_remove}' from {file_path}.")

except FileNotFoundError:
    print(f"Error: The file {file_path} was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
with open("../../data/processed_lectures/lecture3_laplace_transforms/cleaned_text_laplace_transform.txt", encoding="utf-8") as f:
    text_file = f.read()

print(f"Length of text file: {len(text_file)} characters")
```
