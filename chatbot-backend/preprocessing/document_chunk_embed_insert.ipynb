```python
import sys
import os
import glob
sys.path.append(os.path.abspath(".."))
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy.exc import IntegrityError
from db.models import DocumentChunk
from dotenv import load_dotenv
from sentence_transformers import SentenceTransformer
```

```python
load_dotenv()
POSTGRES_USER = os.getenv("POSTGRES_USER")
POSTGRES_PASS = os.getenv("POSTGRES_PASS")
POSTGRES_HOST = os.getenv("POSTGRES_HOST")
POSTGRES_DB = os.getenv("POSTGRES_DB")
POSTGRES_PORT = os.getenv("POSTGRES_PORT")

# Initialise database URL
database_url = f"postgresql://{POSTGRES_USER}:{POSTGRES_PASS}@{POSTGRES_HOST}:{POSTGRES_PORT}/{POSTGRES_DB}"
engine = create_engine(database_url)
Session = sessionmaker(bind=engine)
session = Session()
```

```python
print("Loading BAAI/bge-m3 model...")
embedding_model = SentenceTransformer('BAAI/bge-m3')
print("Model loaded.")
```

```python
# load txt files from data/processed_lectures directory
file_paths = glob.glob("../../data/processed_lectures/**/*.txt", recursive=True)
# process lecture1_intro_to_process_control
file_paths_chapter1 = [fp for fp in file_paths if "lecture1_intro_to_process_control" in fp]
file_paths_chapter1
```

```python
# find the other chapter file paths
file_paths_chapter2 = [fp for fp in file_paths if "lecture2_theoretical_models_chemical_process" in fp]
file_paths_chapter2
```

```python
file_paths_chapter3 = [fp for fp in file_paths if "lecture3_laplace_transforms" in fp]
file_paths_chapter3
```

```python
file_paths_chapter4 = [fp for fp in file_paths if "lecture4_transfer_funct_models" in fp]
file_paths_chapter4
```

```python
file_paths_chapter5 = [fp for fp in file_paths if "lecture5_dynamic_behaviour_first_second_order" in fp]
file_paths_chapter5
```

```python
file_paths_chapter6 = [fp for fp in file_paths if "lecture6_dynamic_response_characteristics" in fp]
file_paths_chapter6
```

```python
# Combine all file paths into a single list while maintaining order
all_file_paths = (
    file_paths_chapter1 +
    file_paths_chapter2 +
    file_paths_chapter3 +
    file_paths_chapter4 +
    file_paths_chapter5 +
    file_paths_chapter6
)

all_file_paths
```

```python
def process_and_insert_chunks_with_incrementing_subtopic_ids(file_paths):
    """
    Process each file, split into paragraph chunks, embed, and insert into the database
    with incrementing subtopic IDs.
    """
    subtopic_id = 1
    for file_path in file_paths:
        with open(file_path, "r", encoding="utf-8") as file:
            content = file.read().strip()
            paragraphs = [p.strip() for p in content.split("\n\n") if p.strip()]  # Split by paragraphs
            
            for paragraph in paragraphs:
                embedding = embedding_model.encode(paragraph).tolist()
                
                document_chunk = DocumentChunk(
                    subtopic_id=subtopic_id,
                    content=paragraph,
                    embedding=embedding
                )
                
                try:
                    session.add(document_chunk)
                    session.commit()
                except IntegrityError:
                    session.rollback()
                    print(f"Failed to insert chunk: {paragraph[:50]}...")
        
        # increment subtopic_id for the next file
        subtopic_id += 1

# process all files with incrementing subtopic IDs
process_and_insert_chunks_with_incrementing_subtopic_ids(all_file_paths)
```

```python
session.close()
```
